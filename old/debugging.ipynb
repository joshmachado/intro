```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import pandas as pd
import os.path
```

```python
tab1 = Table.read('/Users/josh/projects/intro/ngc6300/ngc6300_12m+7m+tp_co21_90pc_props.fits.bz2') #ORIGINAL
tab2 = Table.read('/Users/josh/projects/intro/ngc6300/max/ngc6300_12m+7m+tp_co21_90pc_props.fits.bz2') #FROM ERIK
tab3 = Table.read('/Users/josh/projects/intro/ngc6300/matched/ngc6300_12m+7m+tp_co21_90pc_props.fits.bz2') #ORIGINAL MATCHED

x1 = tab1['XCTR_DEG']
y1 = tab1['YCTR_DEG']
x1 = (x1-np.mean(x1))*np.cos(np.deg2rad(y1))

x2 = tab2['XCTR_DEG']
y2 = tab2['YCTR_DEG']
x2 = (x2-np.mean(x2))*np.cos(np.deg2rad(y2))
x2_max = tab2['XMAX_DEG']
y2_max = tab2['YMAX_DEG']
x2_max = (x2_max-np.mean(x2_max))*np.cos(np.deg2rad(y2_max))

x3 = tab3['XCTR_DEG']
y3 = tab3['YCTR_DEG']
x3 = (x3-np.mean(x3))*np.cos(np.deg2rad(y3))
```

```python
plt.figure(figsize=(12,12))
plt.scatter(x1, y1, c='red', alpha=0.3, label='Homogenized Centers')
plt.scatter(x2, y2, c='blue', alpha=0.3, label='New Table from Erik - Centers')
plt.legend()
plt.title('NGC 6300')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.scatter(x3, y3, c='black', marker='x', label='Original - Matched Table - Centers')
plt.scatter(x2, y2, c='blue', alpha=0.3, label='New Table from Erik - Centers')
plt.legend()
plt.title('NGC 6300')
plt.show()
```
